We start by replace the null values for the file df_cat

First of all we split the StoreId to obtain the town to complete the null values of the State column.
We do the same to split the ProdId to obtain the name and type of product

```python
df_cat [['StoreId_town', 'StoreId_road']] = df_cat ['StoreId'].str.split('-', expand=True)
df_cat [['ProdId_code', 'ProdId_type']] = df_cat ['ProdId'].str.split('_', expand=True)
df_cat
```

```python
df_cat['State'] = df_cat.groupby('StoreId')['State'].transform(lambda x: x.fillna(x.mode().iloc[0]))
df_cat['Position'] = df_cat.groupby('State')['Position'].transform(lambda x: x.fillna(x.mode().iloc[0]))
df_cat['Product'] = df_cat.groupby('ProdId_code')['Product'].transform(lambda x: x.fillna(x.mode().iloc[0]))
df_cat['Type'] = df_cat.groupby('ProdId_type')['Type'].transform(lambda x: x.fillna(x.mode().iloc[0]))
df_cat
```

We have few explanations concerning the calculation of the company's expected (budget). We therefore exclude them from our analysis. 
We also note that we cannot find the profit calculation using the Sales - Total expenses formula. The same applies to margin. We have also removed the Profit and Margin columns.
We will not use the inventory column so we delete it.

```python
del df_num ['Expected Profit']
del df_num ['Expected Sales']
del df_num ['Expected Margin']
del df_num ['Expected COGS']
del df_num ['Profit']
del df_num ['Margin']
del df_num ['Inventory']
df_num
```

We find NaN in Sales for 2 553 values, i.e. 7% of the dataset. We decided to delete these lines to obtain a complete analysis of the data entered on the real dataset.

```python
df_num = df_num.dropna(subset= ['Sales'])
df_num
```

The Total Expenses column is equal to the COGS + Marketing formula, so we replace the Total Expenses NaNs with this formula. Continuing, we replace the COGS and Marketing NaNs with this same formula.

```python
df_num ['Total Expenses'] = df_num ['Total Expenses'].fillna(df_num ['COGS'] + df_num ['Marketing'])
```

```python
df_num ['COGS'] = df_num ['COGS'].fillna(df_num ['Total Expenses'] - df_num ['Marketing'])
df_num ['Marketing'] = df_num ['Marketing'].fillna(df_num ['Total Expenses'] - df_num ['COGS'])
df_num
```

```python
df_num.isnull().sum()
```

We create rate columns to show the % share of COGS and Marketing in Total Expenses.

```python
df_num.loc[:, 'TxCOGS'] = df_num['COGS'] / df_num['Total Expenses']
df_num.loc[:, 'TxMarketing'] = df_num['Marketing'] / df_num['Total Expenses']
df_num
```

We use .describe() to find out the average rate.

```python
df_num.describe()
```

We can see that the average for COGS is 71% and the average for Marketing is 29%. We use these rates to complete the NaN values in the COGS column (287 or 0.85% of the remaining dataset) and Marketing (297 or 0.88% of the remaining dataset).

```python
df_num['COGS'] = df_num['COGS'].fillna(df_num['Total Expenses'] * 0.71)
df_num['Marketing'] = df_num['Marketing'].fillna(df_num['Total Expenses'] * 0.29)
df_num
```

```python
df_num.isnull().sum()
```

We perform a conditional calculation to fill the NaN of Total expenses according to the averages of the calculated COGS and Marketing rates.

```python
df_num['Total Expenses'] = df_num.apply(lambda row: row['Total Expenses'] if not pd.isna(row['Total Expenses']) else (row['COGS'] / 0.71 if not pd.isna(row['COGS']) else (row['Marketing'] / 0.29 if not pd.isna(row['Marketing']) else np.nan)), axis=1)
df_num
```

```python
df_num.isnull().sum()
```

We repeat our calculation to fill in NaN COGS and Marketing.

```python
df_num['COGS'] = df_num['COGS'].fillna(df_num['Total Expenses'] * 0.71)
df_num['Marketing'] = df_num['Marketing'].fillna(df_num['Total Expenses'] * 0.29)
df_num
```

```python
df_num.isnull().sum()
```

Next, we need to fill in the NaN Market fields in the df_cat file. We assume that the nature of the market is indicated according to the amount of sales. We therefore need to join the two files.

```python
df_merged = pd.merge(df_cat, df_num, on='ReportId', how='inner')
df_merged
```

```python
sales_by_market = df_merged.groupby('Market')['Sales'].sum()
sales_by_market
```

```python
# We control the Market category in line with sales 
result = df_merged.groupby('Market')['Sales'].agg(['min', 'max']).reset_index()
result.columns = ['Market', 'Min_Sales', 'Max_Sales']
result
```

The Medium Market category starts at 71. However, we can see that the Small Market is between 119 and 213. We therefore need to switch the Medium Market from 71 to 118 to Micro Market, and from 119 to 213 to Small Market.

```python
# Function to classify markets
def classify_market(sales):
    if sales == 0:
        return "Irrelevant Market"
    elif 1 <= sales <= 118:
        return "Micro Market"
    elif 119 <= sales <= 213:
        return "Small Market"
    elif 214 <= sales <= 485:
        return "Medium Market"
    elif sales > 485:
        return "Big Market"

df_merged['Market'] = df_merged['Sales'].apply(classify_market)

# The new DataFrame will be called df_mergedmarketfinal
df_mergedfinal = df_merged
```

```python
df_mergedfinal
```

We control the application of the function

```python
resultfinal = df_mergedfinal.groupby('Market')['Sales'].agg(['min', 'max']).reset_index()
resultfinal.columns = ['Market', 'Min_Sales', 'Max_Sales']
resultfinal
```
